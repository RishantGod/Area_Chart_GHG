```python
import pandas as pd

df = pd.read_csv('data.csv')
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df = df.iloc[:-3]
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Grand Total', data=df)
```

```python
df_new = df.drop(columns=['Grand Total'])
df_new
```

```python
df_new.describe()
```

```python
df_new.corr()
```

```python
df_new.corr().style.background_gradient(cmap='coolwarm')
```

```python
df_melted = df_new.melt(id_vars=['Year'], value_vars=['drought', 'flood', 'storm', 'extreme_temperature','wildfire'])
```

```python
df_melted.info()
```

```python
df_melted['Year'] = pd.to_datetime(df_melted['Year'], format='%Y')
df_melted.info()
```

```python
df_melted
```

```python
plt.figure(figsize=(14, 8))
sns.lineplot(data=df_melted, x='Year', y='value', hue='variable')
plt.title('Line Charts for Different Variables Over Years')
plt.xlabel('Year')
plt.ylabel('Value')
plt.legend(title='Variable')
plt.show()
```

```python
df_melted.to_csv('data_cleaned.csv', index=False)
```
